```python
# Importing relevent data
df = pd.read_csv('Divvy_Trips_2019.csv')

# Removing the previous index column because it is full of duplicates
df.drop(columns=['index'], inplace=True)
```

```python
# Some columns are not their correct data type
# This will be fixed to make data cleaning/analysis easier
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

df['tripduration'] = df['tripduration'].str.replace(',','')
df['tripduration'] = pd.to_numeric(df['tripduration'])
```

```python
# Checking for null/missing data
df.isnull().sum()
```

The missing or null values make up a significant portion of the dataset, making it impractical to remove without risking significant information loss. These missing/null data do not stem from data entry errors but rather from instances where riders have chosen not to provide their birth year or gender information. It's worth noting that this missing data is limited to the `gender` and `birthyear` columns and does not impact the integrity of other columns or variables in the dataset. Therefore, it is safe to leave the missing/null data alone.

```python
# Checking for any duplicated entries
df.duplicated().sum()
```

```python
# Creating a boxplot to look if there is a skew in tripduration
plt.boxplot(x=df['tripduration'], showfliers=False, vert=False)

plt.title("Box Plot of Trip Duration")
plt.xlabel("Trip Duration")

plt.show()
```

```python
# Looking at the distribution of tripduration to see if
# it follows a normal distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

df['tripduration'].hist(ax=axes[0])
sm.qqplot(df['tripduration'], line='s', ax=axes[1])

axes[0].set_title("Histogram of Trip Duration")
axes[1].set_title("Q-Q Plot of Trip Duration")

plt.tight_layout()
plt.show()
```

The dataset appears to be free of errors. However, the presence of outliers
will make any insights from the dataset innaccurate. The outliers
will have to be addressed.

```python
# Getting quartiles to identify outliers

# Getting quartiles from tripduration since it is not a normal distribution
Q1 = df['tripduration'].quantile(0.25)
Q3 = df['tripduration'].quantile(0.75)

# Calculating interquartile range
IQR = Q3 - Q1

# Defining the lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
higher_bound = Q3 + 1.5 * IQR

# Getting the outliers
outlier_mask = (df['tripduration'] < lower_bound) | (df['tripduration'] > higher_bound)
iqr_outliers = df[outlier_mask]
iqr_outliers = iqr_outliers.copy()

# Seeing if the outliers make up a sizeable amount of the dataset
iqr_outliers.shape[0]/df.shape[0]*100
```

```python
# Getting a look at the outliers
df[outlier_mask].sort_values(by='tripduration', ascending=False)
```

7% of the dataset is a notable percentage, but given the size of the dataset as a whole, it is not enough to cause a significant loss of information. Furthermore, the heavy skew the outliers cause justifies the elimination of this subset of data.

```python
# Removing these outliers
outlier_indexes = iqr_outliers.index
df.drop(outlier_indexes, inplace=True)

# Seeing if removal was successful
df.shape
```

```python
# Looking at the distribution of tripduration to see if
# it now follows a normal distribution after dropping outliers
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

df['tripduration'].hist(ax=axes[0])
sm.qqplot(df['tripduration'], line='s', ax=axes[1])

axes[0].set_title("Histogram of Trip Duration")
axes[1].set_title("Q-Q Plot of Trip Duration")

plt.tight_layout()
plt.show()
```

```python
# Since the plots now display a normal distribution, I will then confirm that 
# there are no outliers by using the empirical rule
trip_mean = df['tripduration'].mean()
trip_std = df['tripduration'].std()

# Getting the lower and upper limit
trip_low_lim = trip_mean - 3 * trip_std
trip_high_lim = trip_mean + 3 * trip_std

# Getting the outliers (if any)
mask = (df['tripduration'] < trip_low_lim) | (df['tripduration'] > trip_high_lim)
df[mask].head(10)
```

```python
# Seeing how much the remaining outliers make up the dataset
# to see if it will be safe to remove them
df[mask].shape[0]/df.shape[0]*100
```

Although this is not really necessary, I believe it is always best to clean the dataset as much as you can in order to receive the best and most accurate insights. Furthermore, these outliers only make up 0.67% of the dataset making it unlikely to cause loss of information.

```python
# Removing the remaining outliers
outlier_indexes = df[mask].index
df.drop(outlier_indexes, inplace=True)

df.reset_index(inplace=True)
df.drop(columns=['index'], inplace=True)

# Confirming that removal was successful
df.shape
```

```python
# Introducing a "Minutes" column to improve the clarity of trip durations,
# as they are currently recorded in seconds.
df['minutes'] = df['tripduration']/60

# Confirming the column was created
df.head()
```

```python
# Saving the revised & cleaned dataset in "Divvy_Trips_2019_cleaned.csv"
df.to_csv("Divvy_Trips_2019_cleaned.csv", index=False)
```
