Some certain visualizations cannot be made in Tableau because of their complicated nature, so instead I will be creating them in Python.

```python
# Importing relevant packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import datetime as dt
```

```python
# Going to the directory where the necessary Data is located
data_path = os.getcwd() + "\\Documents\\ProjectsWRK\\Datasets\\Divvy_Trips_Capstone_Folder"
try:
    os.chdir(data_path)
except:
    pass

# Confirming the directory is correct
os.listdir()
```

```python
# Importing the cleaned dataset
df = pd.read_csv('Divvy_Trips_2019_cleaned.csv')

df.head()
```

```python
# Visualizing the differences in average Trip Duration between each Rider Type
plt.figure(figsize=(4, 6))
avg_duration = df.groupby('usertype')['minutes'].mean().reset_index()

sns.barplot(x=avg_duration['usertype'], y=avg_duration['minutes'])

plt.xlabel("Rider Type")
plt.ylabel("Avg. Trip Duration (minutes)")
plt.title("Average Trip Duration per Rider Type")

plt.show()
```

```python
# Converting necessary columns to datetime in order to
# add date related columns for later analysis
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

df.info()
```

```python
# Adding a day column
df['day'] = df['start_time'].dt.strftime('%j')

df.head()
```

```python
# Finding the daily average trips taken per Rider type
trips_per_day = df.groupby(['day', 'usertype'])['trip_id'].count()
avg_daily_trips = trips_per_day.groupby('usertype').mean().reset_index()

avg_daily_trips
```

```python
# Visualizing the Daily average rides by Rider Type
plt.figure(figsize=(4, 6))
sns.barplot(x=avg_daily_trips['usertype'], y=avg_daily_trips['trip_id'])

plt.xlabel("Rider Type")
plt.ylabel("Trips Taken")
plt.title("Daily Average rides by Rider Type")

plt.show()
```

```python
# Adding a "Weekday" column
df['weekday'] = df['start_time'].dt.strftime('%a')

df.head()
```

```python
# Finding the average trips taken for each day of the week
trips_per_day = df.groupby(['weekday', 'day', 'usertype'])['trip_id'].count().reset_index()
custom_sort_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

avg_weekday_trips = trips_per_day.groupby(['weekday', 'usertype'])['trip_id'].mean().reset_index()

# Sorting the results
avg_weekday_trips['weekday'] = pd.Categorical(avg_weekday_trips['weekday'], categories=custom_sort_order, ordered=True)
avg_weekday_trips.sort_values(by='weekday', inplace=True)

avg_weekday_trips
```

```python
# Visualizing the Average Trips per Day of the Week
sns.barplot(data=avg_weekday_trips, x='weekday', y='trip_id', hue='usertype');

plt.xlabel("Day of the Week")
plt.ylabel("Avg. Trips Taken")
plt.title("Avg. Trips per Day of the Week")

plt.legend(title="Rider Type", loc="center left", bbox_to_anchor=(1.0, 0.5))

plt.show()
```

```python
# Finding the average amount of trips taken daily for each Station
trips_per_day = df.groupby(['day', 'from_station_name', 'from_station_id', 'usertype'])['trip_id'].count().reset_index()
station_trips = trips_per_day.groupby(['from_station_name', 'from_station_id', 'usertype'])['trip_id'].mean().reset_index()

# Sorting the results from greatest to least
top_stn_trips = station_trips[station_trips['usertype'] == 'Customer'].sort_values(by='trip_id', ascending=False)
top_stn_trips.head(20)
```

```python
# Rearranging the columns
top_stn_trips = top_stn_trips[['from_station_id', 'from_station_name', 'usertype', 'trip_id']]

# Dropping the "usertype" column as it is unnecessary
top_stn_trips.drop(columns='usertype', inplace=True)

# Renaming the columns for added clarity
top_stn_trips.columns = ["Station ID", "Station Name", "Daily Trips (Average)"]

top_stn_trips.head(5)
```

```python
# Getting the top 10 stations with the highest daily trips
top_10_stn = top_stn_trips.head(10)

top_10_stn
```

```python
# Getting the average amount of daily trips taken for each station
trip_avg = top_stn_trips["Daily Trips (Average)"].mean()

# Visualizing the Top 10 stations' Average Trip Count in comparison to the average Trip Count
# for all stations
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

sns.barplot(x=top_10_stn["Station Name"], y=top_10_stn["Daily Trips (Average)"])

plt.axhline(y=trip_avg, color='red', linestyle='--', linewidth=2, label="Average Daily Trips (All stations)")

plt.legend()

plt.xticks(rotation=45, fontsize=9)

plt.xlabel("Station Name")
plt.ylabel("Daily Trips (Average)")
plt.title("Top 10 Stations' Daily Average Trips")

plt.tight_layout()
plt.show()
```

```python
# Adding a column that shows how many times greater a station's average trips are than the overall average
# This is so Stakeholders can easily pinpoint which Stations attract a lot of Casual Riders
top_stn_trips['Times Greater than Average'] = (top_stn_trips['Daily Trips (Average)']/trip_avg).round(2).astype(str) + 'x'

top_stn_trips
```

```python
# Saving the Station information to a CSV file to share with Stakeholders
top_stn_trips.to_csv("Station_Trip_INFO.csv", index=False)
```
